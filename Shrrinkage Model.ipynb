```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Load the dataset
dataset = pd.read_csv(r'C:\Users\Hassan Fathy\Downloads\shrink_ray_dataset.csv')
print(dataset.head())
```

```python
X = dataset['knob'].values.reshape(-1, 1)
y = dataset['Shrinkage'].values
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
# Scatter plot of the dataset
plt.scatter(X, y, color='red', label='Dataset')

# Plot the linear regression line
plt.plot(X, model.predict(X), color='blue', linewidth=2, label='Linear Regression')

# Add labels and title to the plot
plt.xlabel('Knob')
plt.ylabel('Shrinkage')
plt.title('Linear Regression')
plt.legend()

# Display the plot
plt.show()
```

```python
# Calculate the knob adjustment for 85% shrinkage
desired_shrinkage = 85
knob_adjustment = (desired_shrinkage - model.intercept_) / model.coef_

print(f"The knob should be adjusted to: {knob_adjustment[0]:.2f} MW")
```
